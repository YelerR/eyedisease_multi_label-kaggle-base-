# 폴더 path

```python
#N_save_path = 'preprocessed_images(divide)/N' 그대로 유지
D_path = 'preprocessed_images(divide)/D'
O_path = 'preprocessed_images(divide)/O'
C_path = 'preprocessed_images(divide)/C'
G_path = 'preprocessed_images(divide)/G'
A_path = 'preprocessed_images(divide)/A'
M_path = 'preprocessed_images(divide)/M'
H_path = 'preprocessed_images(divide)/H'
ex_path = 'preprocessed_images(divide)/ex'
```

```python
def center_crop(img, set_size):

    h, w, c = img.shape

    if set_size > min(h, w):
        return img

    crop_width = set_size
    crop_height = set_size

    mid_x, mid_y = w//2, h//2
    offset_x, offset_y = crop_width//2, crop_height//2
       
    crop_img = img[mid_y - offset_y:mid_y + offset_y, mid_x - offset_x:mid_x + offset_x]
    return crop_img
```

# C 데이터 증강
- 293 -> 2000
- 90 / 180/ 270도 회전
- center crop
- medianBlur

```python
C_i = 0
for C_i in range(len(C_name)):
    C_img = cv2.imread(C_path + '/' + C_name[C_i], cv2.IMREAD_COLOR)
    C_img90 = cv2.rotate(C_img, cv2.ROTATE_90_CLOCKWISE) # 시계방향으로 90도 회전
    C_img180 = cv2.rotate(C_img, cv2.ROTATE_180) # 180도 회전
    C_img270 = cv2.rotate(C_img, cv2.ROTATE_90_COUNTERCLOCKWISE) #  시계방향으로 270도 회전
    C_img_crop = center_crop(C_img, 370)
    C_img_blur = cv2.medianBlur(C_img,11)
    
    cv2.imwrite(os.path.join(C_path , f'C_90_' + C_name[C_i]), C_img90)
    cv2.imwrite(os.path.join(C_path , f'C_180_' + C_name[C_i]), C_img180)
    cv2.imwrite(os.path.join(C_path , f'C_270_' + C_name[C_i]), C_img270)
    cv2.imwrite(os.path.join(C_path , f'C_crop_' + C_name[C_i]), C_img_crop)
    cv2.imwrite(os.path.join(C_path , f'C_blur_' + C_name[C_i]), C_img_blur)
```
