```python
import cv2
import glob
import os
```

```python
a = glob.glob('preprocessed_images/*')
```

```python
lo = []
i = 0
for i in range(len(a)):
    lo.append(a[i].split('/')[1])
    print(lo)
```

```python
lo
```

```python


load_path = 'preprocessed_images'
save_path = 'preprocessed_images(channel)'


for li in range(len(lo)):
    img = cv2.imread(load_path + '/' + lo[li], cv2.IMREAD_COLOR)

    b, g, r = cv2.split(img)
   
    inverse = cv2.merge((r, g, b))

    cv2.imwrite(os.path.join(save_path , lo[li] ), inverse)
```
